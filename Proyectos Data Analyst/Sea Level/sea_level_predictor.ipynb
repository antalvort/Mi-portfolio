```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('epa-sea-level.csv')
```

```python
df
```

```python
df.info()
```

```python
x = df[['Year']]
y = df['CSIRO Adjusted Sea Level']
```

```python
plt.scatter(x,y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
prediccionlr=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
fiabilidad=r2_score(y_test,prediccionlr)
```

```python
fiabilidad
```

```python

fecha_nueva = np.array([[2050]])
prediccion = lr.predict(fecha_nueva)

```

```python
prediccion
```

```python
# Graficar los datos de prueba y la predicciÃ³n
plt.figure(figsize=(10, 6))
plt.scatter(x_test, y_test, color='blue', label='Datos de Prueba')  # Graficar los datos de prueba
plt.plot(x_test, prediccionlr, color='red', linewidth=2, label='PredicciÃ³n')  # Graficar la predicciÃ³n

# Etiquetas y tÃ­tulo del grÃ¡fico
plt.xlabel('AÃ±o')
plt.ylabel('CSIRO Adjusted Sea Level')
plt.title('PredicciÃ³n del Nivel del Mar con RegresiÃ³n Lineal')
plt.legend()

plt.show()
```
